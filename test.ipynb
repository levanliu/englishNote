```python
import numpy as np
import matplotlib.pyplot as plt

# Read data from files
x, f = np.loadtxt('plot_data.txt', unpack=True)
x, fFS = np.loadtxt('fourier_data.txt', unpack=True)

# Plot original function
plt.plot(x, f, '-', color='k', linewidth=2, label='Original')

# Plot Fourier series approximation
plt.plot(x, fFS, '-', label='Fourier Series')

# Set plot properties
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Fourier Series Approximation')
plt.legend()

# Show the plot
plt.show()
```

```python
import sympy as sp

# Define the variable
x = sp.Symbol('x')

# Define the function
f = 1/(1+x**3)

# Find the antiderivative
F = sp.integrate(f, x)

# Display the result
print(F)
```

```python
import sympy as sp
import numpy as np
import math

# 定义变量和函数
x = sp.Symbol('x')
# f = x**2 + 3*x + 2
f = (sp.log(1+x))/(x*sp.sqrt(1+x))
# 计算定积分
a = 0
b = 1
result = sp.integrate(f, (x, a, b))

# 打印结果
print(result)
```
